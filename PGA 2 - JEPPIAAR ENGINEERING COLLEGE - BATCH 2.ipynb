**QUESTION 4**



**Annual project of university done by three groups of students with equal sample sizes. Each group was given a different task. After the final presentation students get the marks. The summary statistics are given below:
Calculate the F statistics**
image.png

---

```python
import numpy as np

mean1 = 44.5
mean2 = 42
mean3 = 46.5

#sample size
n = 6

std_dev1 = 5
std_dev2 = 5
std_dev3 = 8

# Calculate the total mean
total_mean = (mean1 + mean2 + mean3) / 3

# Calculate the "Between-group variance"
between_group_variance = (n * (mean1 - total_mean)**2 +
                          n * (mean2 - total_mean)**2 +
                          n * (mean3 - total_mean)**2) / (3 - 1)

# Calculate the "Within-group variance"
within_group_variance = ((n - 1) * std_dev1**2 +
                        (n - 1) * std_dev2**2 +
                        (n - 1) * std_dev3**2) / (15 - 3)

# Calculate the F-statistic
F = between_group_variance / within_group_variance

print("F-statistic:", F)
```

**QUESTION 8**

**The number of deaths in the 7 metro cities of the US during the last month due to Covid-19 is recorded by the government, and that time, experts predicted that upto next month death rate will be increased by 30%. So, the number of deaths in current month is also recorded in 7 cities in US are given below:
City = ["New York", "New Jersey", "Michigan", "California", "Florida", "Massachusetts", "Texas"] No_of_deaths_in_last_month = [3406, 1469, 662, 583, 582, 526, 461] No_of_deaths_in_current_month = [4398, 1846, 1288, 382, 879, 430, 321]
df = pd.DataFrame({"City": City, "No_of_deaths_in_last_month":No_of_deaths_in_last_month, "No_of_deaths_in_current_month": No_of_deaths_in_current_month}) df
Is the death distribution of the current month the same as the experts prediction? Use the level of significance is 0.1.**

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Data Setup
data = {
    "City": ["New York", "New Jersey", "Michigan", "California", "Florida", "Massachusetts", "Texas"],
    "Deaths_Last_Month": [3406, 1469, 662, 583, 582, 526, 461],
    "Deaths_Current_Month": [4398, 1846, 1288, 382, 879, 430, 321]
}

# Creating DataFrame
df = pd.DataFrame(data)

# Calculate expected deaths with a 30% increase over last month
df["Expected_Deaths"] = df["Deaths_Last_Month"] * 1.3

# Chi-square test for observed vs. expected deaths
observed = df["Deaths_Current_Month"]
expected = df["Expected_Deaths"]
chi2, p, dof, expected_values = chi2_contingency([observed, expected])

# Significance level
alpha = 0.1

# Hypothesis Testing
if p < alpha:
    print("Reject the null hypothesis: The death distribution of the current month is different from the expert's prediction.")
else:
    print("Fail to reject the null hypothesis: The death distribution of the current month is not significantly different from the expert's prediction.")
```

### c) Dealing with correlated variables
------------

```python
data = data.drop(columns=['CUST_ID']) # Drops the 'CUST_ID' column if present
```

```python
plt.figure(figsize = (13,8))
sns.heatmap(data.corr(), annot = True)
```

```python
import pandas as pd

# Calculate the correlation matrix
correlation_matrix = data.corr().abs()

# Set an upper triangle of the correlation matrix to avoid duplicate pairs
upper_triangle = correlation_matrix.where(
    np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool)
)

# Find columns with a correlation higher than the threshold (e.g., 0.8)
threshold = 0.8
to_drop = [column for column in upper_triangle.columns if any(upper_triangle[column] > threshold)]

# Drop the identified columns
data_reduced = data.drop(columns=to_drop)

print("Dropped columns due to high correlation:", to_drop)
```

* We will remove columns that have correlations above the specified threshold, it will simplify DataFrame by eliminating multicollinearity.
